```python
def import_fasttext_vectors(path, limit=None):
    file = open(path, 'r') 
    lines = file.readlines()
    length = int(lines[0].split()[0])
    x, y = [], []
    
    if limit and limit < length:
        length = limit
        
    for i in range(1, length):
        x.append(lines[i].split()[1:])
        y.append(lines[i].split()[0])
        
    return x, y

x,y = import_fasttext_vectors("../../models/enwik_e15_d200.vec", limit=1000)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
```

```python
X_2d = tsne.fit_transform(x)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(25, 60))

x_0 = [x[0] for x in X_2d]
x_1 = [x[1] for x in X_2d]

plt.scatter(x_0, x_1, c='r', label='enwik_e15_d200')


for i in range(len(X_2d)):
    plt.annotate(y[i], (x_0[i], x_1[i]), fontsize=12)


plt.legend()
plt.show()
```
